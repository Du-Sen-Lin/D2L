# Attention Pooling: Nadaraya-Watson Kernel Regression
概括地说，查询（意志线索）和键（非意志线索）之间的相互作用导致注意力集中。注意池有选择地聚合值（感官输入）以产生输出。

1964 年提出的 Nadaraya-Watson 核回归模型是一个简单而完整的例子，用于演示具有注意力机制的机器学习。

```python
import torch
from torch import nn
from d2l import torch as d2l
import sys
sys.path.append('../../CommonFunctions/tools/')
import common_tools
import sequence_tools
import attention_tools
```

## 1、生成数据集

```python
n_train = 50
x_train, _ = torch.sort(torch.rand(n_train) * 5)

def f(x):
    return 2 * torch.sin(x) + x**0.8

y_train = f(x_train) + torch.normal(0.0, 0.5, ((n_train,)))
x_test = torch.arange(0, 5, 0.1)
y_truth = f(x_test)
n_test = len(x_test)
n_test
```

以下函数绘制了所有训练示例（用圆圈表示）、f不含噪声项的真实数据生成函数（用“Truth”标记）和学习的预测函数（用“Pred”标记）。

```python
def plot_kernel_reg(y_hat):
    common_tools.plot(x_test, [y_truth, y_hat], 
                      'x', 'y', 
                      legend=['Truth', 'Pred'], 
                      xlim=[0, 5], ylim=[-1, 5])
    common_tools.plt.plot(x_train, y_train, 'o', alpha=0.5)
```

## 2、平均池化

```python
y_hat = torch.repeat_interleave(y_train.mean(), n_test)
plot_kernel_reg(y_hat)
```

## 3、非参数注意力池


x  是查询和  (xi,yi) 是键值对;
Nadaraya-Watson 核回归是一个非参数模型;

我们绘制了基于此非参数注意力模型的预测。与平均池化产生的线相比，预测线平滑且更接近真实情况。

```python
X_repeat = x_test.repeat_interleave(n_train).reshape((-1, n_train))
attention_weights = nn.functional.softmax(-(X_repeat - x_train)**2 / 2, dim=1)
y_hat = torch.matmul(attention_weights, y_train)
plot_kernel_reg(y_hat)
```

注意力权重: 这里测试输入是查询，而训练输入是键。由于两个输入都已排序，我们可以看到查询键对越接近，注意力池中的注意力权重就越高。

```python
attention_tools.show_heatmaps(attention_weights.unsqueeze(0).unsqueeze(0), 
                              xlabel = 'Sorted training inputs', 
                              ylabel='Sorted testing inputs')
```
