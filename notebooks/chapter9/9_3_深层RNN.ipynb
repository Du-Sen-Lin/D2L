## 2、实现

```python
import torch
from torch import nn
from d2l import torch as d2l
import sys
sys.path.append('../../CommonFunctions/tools/')
import common_tools
import sequence_tools
```

```python
batch_size, num_steps = 32, 35
train_iter, vocab = sequence_tools.load_data_time_machine(batch_size, 
                                                          num_steps)
```

```python
vocab_size, num_hiddens, num_layers = len(vocab), 256, 2
num_inputs = vocab_size
device = common_tools.try_gpu(2)
lstm_layer = nn.LSTM(num_inputs, num_hiddens, num_layers)
model = sequence_tools.RNNModel(lstm_layer, len(vocab))
model = model.to(device)
```

## 3、训练与预测

```python
num_epochs, lr = 500, 2
sequence_tools.train_ch8(model, train_iter, vocab, 
                         lr, num_epochs, device)
```
