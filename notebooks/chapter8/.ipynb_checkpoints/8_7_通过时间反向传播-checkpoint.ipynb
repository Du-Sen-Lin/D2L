# 通过时间反向传播
深入地探讨序列模型反向传播的细节，以及相关数学原理。

循环神经网络中的前向传播相对简单。 通过时间反向传播（Backpropagation through time，BPTT） [Werbos, 1990] 实际上是在循环神经网络中应用的一个特定的反向传播技术。


## 1、循环神经网络的梯度分析

可以使用链式法则递归地计算  ∂ht/∂wh ，但当  t  很大，这个链就会变得很长。

### 完整计算
梯度可能会爆炸，因为初始条件的微妙变化可能会对结果产生很大影响。

### 截断时间步
  τ  步后截断求和，这会带来真实梯度的近似，只需将求和终止为 ∂ht−τ/∂wh 。它通常被称为通过时间截断反向传播。
